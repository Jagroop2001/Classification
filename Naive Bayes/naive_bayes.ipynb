# Decision Tree Classifier

## Importing Relevant Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## Reading the data set

```python
data_set = pd.read_csv('Social_Network_Ads.csv')
data_set.head()
```

## Split the data into training and test set

```python
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=0)
```

## Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_train)
```

## Training Logistic Regression model on Training Set 

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB()
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict([[22,16000]])
```

```python
classifier.predict_proba([[22,16000]])
```

```python
classifier.predict(sc.transform([[22,16000]]))
```

## Predict the Test Results 

```python
y_pred = classifier.predict(X_test)
np.set_printoptions(precision=2)
result = print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
print(result)
```

## Confusion Matrix 

```python
from sklearn.metrics import confusion_matrix
cm =confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
